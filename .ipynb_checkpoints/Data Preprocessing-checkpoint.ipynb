```python
import pandas as pd
```

```python
data = pd.read_csv('Titanic.csv')
```

```python
data.info()
```

```python
data1 = data.drop('Name',axis=1)
data1.drop('Ticket',axis=1,inplace=True)
data1.drop('Cabin',axis=1,inplace=True)
data1.head()
```

```python
data1.isnull().sum()
```

```python
data1= pd.get_dummies(data1, columns = ['Sex', 'Embarked'])
data1
```

```python
c = data1.pop('Survived')
data1.insert(len(list(data1)), 'Class', c)
data1
```

```python
titanic_gender = data1[["Sex"]]
```

```python
from sklearn.preprocessing import OneHotEncoder
gender_encoder = OneHotEncoder()
titanic_gender1 = gender_encoder.fit_transform(titanic_gender)
titanic_gender1
```

```python
data3 = pd.DataFrame(titanic_gender1.toarray())
data3
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
titanic_gender2 = sc.fit_transform(titanic_gender1)
```
