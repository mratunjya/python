```python
import pandas as pd
import numpy as np 
```

```python
data = pd.read_csv('smoke_detection_iot.csv')
```

```python
data.head()
```

```python
data.drop(columns=['Index'],inplace=True)
```

```python
data
```

```python
target = data['Fire Alarm']
data1 = data.drop(columns=['Fire Alarm'])
```

```python
print(data1.shape)
print(target.shape)
```

```python
x_train,x_test,y_train,y_test = data1[50104:],data1[:50104],target[50104:],target[:50104]
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd_clf = SGDClassifier(random_state=42)
```

```python
sgd_clf.fit(x_train,y_train)
```

```python
result = sgd_clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,result)
print("Accuracy Score: %0.2f"%(accuracy*100))
```
